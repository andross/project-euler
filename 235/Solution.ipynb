## Solution

First, simplify the expression using Wolfram Alpha, to remove the sum.
Then, substitute 5000. The initial $r_0$ was found using Newton's method with the AD package in Haskell, but precision is too low.

```python
r0 = 1.0023221086328762

def expr(r):
    return -(3*(5000*r**(5000+1) \
              - 300*r**(5000+1) \
              -  5000*r**5000 \
              +  299 * r**5000  \
              +  300 * r \
           - 299))/(r-1)**2  \
           +  600000000000
```

```python
expr(r0) # Not quite zero.
```

```python
from mpmath import mp, mpf, findroot, nstr

mp.dps = 1000
r = findroot(expr, r0)
print "r = ", r

print "remainder = ", expr(r)

print "Rounded to 12 decimals, r = ", nstr(r, n=12)
```
